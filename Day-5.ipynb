```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
x = [1,2,3,4,5]
y = [10,20,25,30,40]

# 1. Style
sns.set_style("whitegrid")

plt.plot(x, y)
plt.title("Styled Plot")
plt.show()


# 2. Figure size
plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.show()


# 3. Annotations
plt.plot(x, y)
plt.annotate("Peak", (5,40))
plt.show()


# 4. Grid
plt.plot(x, y)
plt.grid(True)
plt.show()


# 5. Full customization
plt.figure(figsize=(8,5))
plt.plot(x, y, color='green', marker='o')
plt.title("Final Styled Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()
```
