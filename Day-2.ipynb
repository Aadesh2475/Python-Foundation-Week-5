```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Histogram
data = np.random.randint(50, 100, 100)

plt.hist(data, bins=10)
plt.title("Exam Score Distribution")
plt.show()


# 2. Histogram with bins
plt.hist(data, bins=5)
plt.title("Less bins")
plt.show()


# 3. Scatter Plot
height = [150, 160, 170, 180, 190]
weight = [50, 60, 65, 80, 90]

plt.scatter(height, weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()


# 4. Colored Scatter
plt.scatter(height, weight, c='red')
plt.show()


# 5. Sized Scatter
sizes = [20, 50, 80, 100, 200]
plt.scatter(height, weight, s=sizes)
plt.show()
```
