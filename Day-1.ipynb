```python
import matplotlib.pyplot as plt

# 1. Line Plot - Monthly Sales Trend
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [200, 250, 300, 280, 350]

plt.plot(months, sales, marker='o', label="Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.legend()
plt.show()


# 2. Multiple Line Plot
product_A = [100, 150, 200, 180, 220]
product_B = [80, 120, 160, 140, 200]

plt.plot(months, product_A, label="Product A")
plt.plot(months, product_B, label="Product B")
plt.legend()
plt.title("Product Comparison")
plt.show()


# 3. Bar Chart
products = ["A", "B", "C"]
values = [50, 70, 40]

plt.bar(products, values)
plt.title("Product Sales")
plt.show()


# 4. Horizontal Bar
plt.barh(products, values)
plt.title("Horizontal Bar Chart")
plt.show()


# 5. Customized Bar Chart
plt.bar(products, values, color=['red','blue','green'])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Customized Bar Chart")
plt.show()
```
